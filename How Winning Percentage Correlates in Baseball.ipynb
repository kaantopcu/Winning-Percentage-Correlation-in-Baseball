```python
# Importing necessary packages
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
# Importing team level stats of MLB from season 2010 to 2022
allStats = pd.read_csv('AllSeasonsStats.txt')
allStats
```

```python
# Listing all the columns in the df
print(allStats.columns.tolist())
```

```python
# Creating a binary variable defines the result as a home team win or visitor team win
allStats['hwin'] = np.where(allStats['HomeRunsScore'] > allStats['VisitorRunsScored'],1,0)
allStats['awin'] = np.where(allStats['HomeRunsScore'] < allStats['VisitorRunsScored'],1,0)
```

```python
# Extracting the year from the first 4 digit of the date column
allStats['year'] = allStats['Date'].astype(str).str[0:4]
allStats.head()
```

```python
# Aggregating the team's stats for the games in which they are home team

Teamshome = allStats.groupby(['HomeTeam','year'])['VisitorAB', 'VisitorH', 'VisitorD', 'VisitorT', 'VisitorHR', 'VisitorSF',
                                           'VisitorHBP','VisitorBB','VisitorK','HomeAB', 'HomeH', 'HomeD', 'HomeT',
                                           'HomeHR', 'HomeSF', 'HomeHBP', 'HomeBB','HomeK','hwin','awin'].sum()



Teamshome = Teamshome.reset_index()
Teamshome = Teamshome.rename(columns = {"HomeTeam": "Team"}) 
Teamshome
```

```python
# Aggregating the team's stats for the games in which they are visitor team
Teamsaway = allStats.groupby(['year','VisitingTeam'])['VisitorAB', 'VisitorH', 'VisitorD', 'VisitorT', 'VisitorHR', 'VisitorSF',
                                           'VisitorHBP','VisitorBB','VisitorK','HomeAB', 'HomeH', 'HomeD', 'HomeT',
                                           'HomeHR', 'HomeSF', 'HomeHBP', 'HomeBB','HomeK','hwin','awin'].sum()

Teamsaway = Teamsaway.reset_index()
Teamsaway = Teamsaway.rename(columns = {"VisitingTeam": "Team"}) 
Teamsaway
```

```python
# Merging the away team stats and home team stats
Teams = pd.merge(Teamshome, Teamsaway, on =['Team','year'])

# Total number of wins by each team in every season
Teams['wins'] = Teams['hwin_x']+Teams['awin_y']
Teams
```

```python
# Calculate On Base Percentage for Both For and Against
Teams['OBPFOR'] = ((Teams['HomeH_x']+Teams['VisitorH_y']+
                    Teams['HomeBB_x']+Teams['VisitorBB_y']+
                    Teams['HomeHBP_x']+Teams['VisitorHBP_y'])/
                    (Teams['HomeAB_x']+Teams['VisitorAB_y']+
                     Teams['HomeBB_x']+Teams['VisitorBB_y']+
                     Teams['HomeHBP_x']+Teams['VisitorHBP_y']+
                     Teams['HomeSF_x']+Teams['VisitorSF_y']))


Teams['OBPAGN'] = ((Teams['HomeH_y']+Teams['VisitorH_x']+
                    Teams['HomeBB_y']+Teams['VisitorBB_x']+
                    Teams['HomeHBP_y']+Teams['VisitorHBP_x'])/
                    (Teams['HomeAB_y']+Teams['VisitorAB_x']+
                     Teams['HomeBB_y']+Teams['VisitorBB_x']+
                     Teams['HomeHBP_y']+Teams['VisitorHBP_x']+
                     Teams['HomeSF_y']+Teams['VisitorSF_x']))
```

```python
# Calculate Slugging Percentage for Both For and Against
Teams['SLGFOR'] = ((Teams['HomeH_x']+Teams['VisitorH_y']-
                    (Teams['HomeD_x']+Teams['VisitorD_y'])-
                    (Teams['HomeT_x']+Teams['VisitorT_y'])-
                    (Teams['HomeHR_x']+Teams['VisitorHR_y'])+
                    2*Teams['HomeD_x']+2*Teams['VisitorD_y']+
                    3*Teams['HomeT_x']+3*Teams['VisitorT_y']+
                    4*Teams['HomeHR_x']+4*Teams['VisitorHR_y'])/
                    (Teams['HomeAB_x']+Teams['VisitorAB_y']))

Teams['SLGAGN'] = ((Teams['HomeH_y']+Teams['VisitorH_x']-
                    (Teams['HomeD_y']+Teams['VisitorD_x'])-
                    (Teams['HomeT_y']+Teams['VisitorT_x'])-
                    (Teams['HomeHR_y']+Teams['VisitorHR_x'])+
                    2*Teams['HomeD_y']+2*Teams['VisitorD_x']+
                    3*Teams['HomeT_y']+3*Teams['VisitorT_x']+
                    4*Teams['HomeHR_y']+4*Teams['VisitorHR_x'])/
                    (Teams['HomeAB_y']+Teams['VisitorAB_x']))
```

```python
# Calculating Batting Percentage for Both For and Against
Teams['BAFOR'] = (Teams['HomeH_x']+Teams['VisitorH_y'])/(Teams['HomeAB_x']+Teams['VisitorAB_y'])
Teams['BAAGN'] = (Teams['HomeH_y']+Teams['VisitorH_x'])/(Teams['HomeAB_y']+Teams['VisitorAB_x'])
```

```python
# Calculating Strike-Outs per Walk for Both For and Against
Teams['KperBBFOR'] = (Teams['HomeK_x']+Teams['VisitorK_y'])/(Teams['HomeAB_x']+Teams['VisitorAB_y'])
Teams['KperBBAGN'] = (Teams['HomeK_y']+Teams['VisitorK_x'])/(Teams['HomeAB_y']+Teams['VisitorAB_x'])
Teams
```

```python
# Keeping the stats to be used while building the regressions
Teams2 = Teams[['year', 'Team', 'wins','OBPFOR','OBPAGN','SLGFOR','SLGAGN','BAFOR','BAAGN','KperBBFOR','KperBBAGN']]
Teams2
```

```python
# Calculating home team games for every team in each season
TeamsGh = allStats.groupby(['year','HomeTeam'])['hwin'].count()
TeamsGh = TeamsGh.reset_index()
TeamsGh = TeamsGh.rename(columns = {"HomeTeam": 'Team'}) 

# Calculating away team games for every team in each season
TeamsGa = allStats.groupby(['year','VisitingTeam'])['awin'].count()
TeamsGa = TeamsGa.reset_index()
TeamsGa = TeamsGa.rename(columns = {"VisitingTeam": 'Team'}) 

# Total number of games played by every team in each season
TeamsG = pd.merge(TeamsGh, TeamsGa, on =['Team','year'])
TeamsG['Games'] = TeamsG['hwin']+TeamsG['awin']
TeamsG = TeamsG[['year','Team','Games']]
TeamsG
```

```python
# Merging two dfs to obtain number of games played by each team in every season
Teams2 = pd.merge(Teams2, TeamsG, on =['Team','year'])
Teams2
```

```python
# Calculating the winning share of teams
Teams2['wpc'] = Teams2['wins']/Teams2['Games']
Teams2
```

```python
# Teams with highest winning share in all seasons 
TopWPC = Teams2.groupby(['Team'])['wins','Games'].sum().reset_index()
TopWPC['wpc'] = TopWPC['wins'] / TopWPC['Games']
TopWPC = TopWPC.sort_values(by='wpc',ascending = False).reset_index(drop=True).head(5)
TopWPC
```

```python
# Top 5 Teams with Highest Win Percentage Since 2010
plt.bar(TopWPC['Team'], TopWPC['wpc'], color ='blue',width = 0.3)
plt.xlabel("Teams")
plt.xticks(rotation=70)
plt.ylabel("Winning Share")
plt.title("Win. Percentages of All Teams since 2010")
plt.grid(True)
plt.show()
```

```python
# Getting every season's team with highest winning percentage
SeasonsBest = Teams2.groupby(['year'])['wpc'].max().reset_index()
SeasonsBest = pd.merge(SeasonsBest,Teams2)
SeasonsBest = SeasonsBest[['year','Team','wpc']]
SeasonsBest.sort_values(by='year',ascending = True)
```

```python
# Every Season's Highest Win Percentage 
plt.plot(SeasonsBest['year'], SeasonsBest['wpc'], color='red', marker='o')
plt.title('Seasons with Highest Win Share', fontsize=14)
plt.xlabel('Season', fontsize=14)
plt.xticks(rotation=45)
plt.ylabel('Winning Percentage', fontsize=14)
plt.grid(True)
plt.show()
```

## Regressions

```python
import statsmodels.formula.api as smf
WinOBP_lm = smf.ols(formula = 'wpc ~ OBPFOR + OBPAGN', data=Teams2).fit()
WinOBP_lm.summary()
```

```python
WinSLG_lm = smf.ols(formula = 'wpc ~ SLGFOR + SLGAGN', data=Teams2).fit()
WinSLG_lm.summary()
```

```python
WinBA_lm = smf.ols(formula = 'wpc ~ BAFOR + BAAGN', data=Teams2).fit()
WinBA_lm.summary()
```

```python
WinKperBB_lm = smf.ols(formula = 'wpc ~ KperBBFOR + KperBBAGN', data=Teams2).fit()
WinKperBB_lm.summary()
```

```python
WinOBPSLG_lm = smf.ols(formula = 'wpc ~ OBPFOR + OBPAGN + SLGFOR + SLGAGN', data=Teams2).fit()
WinOBPSLG_lm.summary()
```

```python
Win_lm = smf.ols(formula = 'wpc ~ OBPFOR + OBPAGN + SLGFOR + SLGAGN + BAFOR + BAAGN + KperBBFOR + KperBBAGN', data=Teams2).fit()
Win_lm.summary()
```

```python
from statsmodels.iolib.summary2 import summary_col
Header = ['1','2','3','4','5','6','']
Table_1 = summary_col([WinOBP_lm,WinSLG_lm,WinBA_lm,WinKperBB_lm,WinOBPSLG_lm,Win_lm],\
                      regressor_order=['Intercept','OBPFOR','OBPAGN','SLGFOR','SLGAGN','BAFOR','BAAGN','KperBBFOR','KperBBAGN'],model_names = Header)
print(Table_1)
```
